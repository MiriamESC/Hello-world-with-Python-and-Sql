```python
# problem 4 list of all kidnapping crimes involving a child
%sql select CASE_NUMBER from CHICAGO_CRIME_DATA where DESCRIPTION = 'CHILD ABDUCTION/STRANGER'
```

```python
# problem 5 crime types recorded at schools
%sql select PRIMARY_TYPE, DESCRIPTION, LOCATION_DESCRIPTION from CHICAGO_CRIME_DATA where LOCATION_DESCRIPTION \
in ('SCHOOL, PRIVATE, BUILDING','SCHOOL, PUBLIC, BUILDING','SCHOOL, PUBLIC, GROUNDS')
```

```python
# problem 7: 5 community areas with highest % of households below poverty line
%sql select COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY from CENSUS_DATA \
order by PERCENT_HOUSEHOLDS_BELOW_POVERTY desc fetch first 5 rows only 
```

```python
# problem 8 community area number with highest recorded crimerate (most prone to crime)
%sql select COMMUNITY_AREA_NUMBER, count(COMMUNITY_AREA_NUMBER) as CRIMENUMBER from CHICAGO_CRIME_DATA \
group by COMMUNITY_AREA_NUMBER order by count(COMMUNITY_AREA_NUMBER) desc limit 1
```

```python
#problem 9 Name of community area with highest hardship index score
%sql select COMMUNITY_AREA_NAME from CENSUS_DATA where HARDSHIP_INDEX = (select max(HARDSHIP_INDEX) from CENSUS_DATA)
```
